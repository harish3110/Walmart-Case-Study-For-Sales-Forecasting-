```python
# Basic imports
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('original_datasets/macro_economic.xlsx')
data
```

```python
data.shape
```

```python
train = pd.read_csv('original_datasets/train.csv')
train.head()
```

```python
train.shape
```

```python
test = pd.read_csv('original_datasets/test.csv')
test.head()
```

```python
# Appending train and test data one over the other = (train + test modified)
df = train.append(test).reset_index(drop=True)
df
```

```python
# Removing 2016 data from econ_data as (train + test modified) doesn't have 2016 data
econ_data_2015 = data[:84]
econ_data_2015
```

```python
# Replicating each row in econ_data thrice so as to match (train + test) data
econ_data_replicated = pd.DataFrame(np.repeat(econ_data_2015.values,3,axis=0))
econ_data_replicated
```

### Null values check on econ_data:

```python
# Checking dataset if there are any null values present
econ_data_replicated.isnull().values.any()
```

*** But the econ_data has columns with special values like '?', which has to be addressed.

```python
# Concatenating (train + test data modified) and econ_data
df_final = pd.concat([df, econ_data_replicated], axis=1)
df_final
```

```python
df_final = pd.DataFrame(df_final)
```

```python
# Adding appropriate column names to df_final
df_final.columns = list(df.columns) + list(econ_data_2015.columns)
```

```python
df_final.T
```

```python
# The final dataset to work with
df_final.T
```

---
## Dropping Unnecessary Features:

```python
# 1. Removing redundant 'Year-Month' column
df_final = df_final.drop('Year-Month', axis=1)
```

```python
# 2. PartyInPower
df_final['PartyInPower'].value_counts()
```

#### Since df_final data is from 2009-2015 and the party in power has been democrats throughout; the columns can be dropped.

```python
# 3. Advertising Expenses

econ_data_2015['AdvertisingExpenses (in Thousand Dollars)'][econ_data_2015['AdvertisingExpenses (in Thousand Dollars)'] == '?'].count()
```

```python
df_final = df_final.drop(['PartyInPower', 'AdvertisingExpenses (in Thousand Dollars)'], axis = 1)
df_final.T
```

#### Since Advertising Expenses is '?' for 73 of the possible 84 values; we can safely eliminate this feature for now. 

#### Further investigation and advice from domain expert is needed to check if data is useful in later model building.
